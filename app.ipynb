```python
import cv2
import numpy as np
from keras.models import load_model

# Load the trained model
model = load_model('Basilfacedetection.h5')

# Set the size of the input images
img_size = (224, 224)

# Class labels
class_labels = ['Another_person','Basil_khan']

def preprocess_image(img):
    """Preprocess the input image to the required size and normalize it."""
    img = cv2.resize(img, img_size)
    img = img.astype('float32') / 255.0
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img

def predict_frame(frame):
    """Predict the class of the input frame and return the predicted class name."""
    img = preprocess_image(frame)
    prediction = model.predict(img)
    predicted_class = class_labels[int(prediction[0][0] > 0.70)]
    return predicted_class

# Initialize the webcam
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break

    # Predict the class of the frame
    predicted_class = predict_frame(frame)

    # Display the frame with the predicted class name
    cv2.putText(frame, predicted_class, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    cv2.imshow('Prediction', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
cap.release()
cv2.destroyAllWindows()
```

```python
#good code
import cv2
import numpy as np
from keras.models import load_model


model = load_model('Basilfacedetection.h5')


img_size = (224, 224)


class_labels = ['Another_person','Basil_khan']

def preprocess_image(img):

    img = cv2.resize(img, img_size)
    img = img.astype('float32') / 255.0
    img = np.expand_dims(img, axis=0)  
    return img

def predict_frame(frame):
    img = preprocess_image(frame)
    prediction = model.predict(img)
    predicted_class = class_labels[int(prediction[0][0] > 0.70)]
    return predicted_class

def detect_faces(frame, face_cascade_path='haarcascade_frontalface_default.xml'):


    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + face_cascade_path)


    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)


    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    return faces


cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    if not ret:
        break


    faces = detect_faces(frame)


    for (x, y, w, h) in faces:

        face = frame[y:y+h, x:x+w]

        predicted_class = predict_frame(face)


        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.putText(frame, predicted_class, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    cv2.imshow('Face Detection and Prediction', frame)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```
